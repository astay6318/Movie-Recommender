```python
def change_id_to_int(df):
    for index, row in df.iterrows():
        try:
            if float('-inf')<=int(row['id']) and int(row['id'])<=float('inf'):
                row['id']=row['id']
        except:
            print(row['id'])
            df=df.drop(index)
    print(df['id'].dtypes)
    df['id']=df['id'].astype(int)
    return df
```

```python
new_df=movies[['id','original_title','genres','belongs_to_collection','overview','tagline','keywords','cast','crew','adult']]
new_df.head()
```

```python
new_df=new_df.replace(np.NaN,'')
new_df.head()
```

```python
new_df['adult']=new_df['adult'].apply(lambda x: "18+" if x=='True' else '')
new_df.adult
new_df['adult'].unique()
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 



```

```python
new_df['genres']=new_df['genres'].apply(convert)
new_df.head()
```

```python
def convert2(text_dict):
    L = []
    try:
        L.append(ast.literal_eval(text_dict)['name'])
    except:
        pass
    print(L)
    return L 
```

```python
print(ast.literal_eval(new_df['belongs_to_collection'][0])['name'])
new_df['belongs_to_collection']=new_df['belongs_to_collection'].apply(convert2)
```

```python
new_df.head()
```

```python
new_df['tagline']=new_df['tagline'].apply(lambda x: x.split())
```

```python
new_df['keywords']=new_df['keywords'].apply(convert)
new_df.head()
```

```python
new_df['cast'][1]
```

```python
def convert3(text):
    L = []
    counter=0
    for i in ast.literal_eval(text):
        if counter<4:
            L.append(i['name'])
            counter=counter+1
        else:
            break
    return L 



```

```python
new_df['cast']=new_df['cast'].apply(convert3)
new_df.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
new_df['crew']=new_df['crew'].apply(fetch_director)
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['genres']+new_df['belongs_to_collection']+new_df['overview']+new_df['tagline']+new_df['keywords']+new_df['cast']+new_df['crew']
new_df['tags']
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python

new_df['tags']=new_df['tags'].apply(collapse)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags']=new_df['tags']+new_df['adult']
```

```python
df=new_df[['id','original_title','tags']]
df.head()
```

```python
df.shape
```

```python
df['tags']=df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('leaf')
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=20000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
df['original_title']=df['original_title'].astype(str)
```

```python
df['original_title']=df['original_title'].apply(lambda x: x.lower())
```

```python
def recommend(movie,n=5):
    try:
        index = df[df['original_title'] == movie.lower()].index[0]
        distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
        for i in distances[1:n+1]:
            print(df.iloc[i[0]].original_title)
    except:
        print("Movie Not in Database")
    
```

```python
recommend('Finding Nemo')
```

```python
recommend('BatMan Begins',10)
```
